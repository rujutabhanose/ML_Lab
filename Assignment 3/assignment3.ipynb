```python
#importing necessary libtaries
%pip install seaborn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("/Users/rujutabhanose/Documents/ML/Lab Assignments/Assignment 2/admission - admission.csv")
```

```python

dataset.head()
```

```python
#bar chart
plt.bar(dataset["University Rating"], dataset["Chance of Admit"], color = 'pink')
plt.xlabel("Chance of Admit")
plt.ylabel("University Rating")
plt.title("University Rating VS Chance of Admit")
plt.show()
```

```python
#scatterplot
plt.scatter(dataset["GRE Score"], dataset["TOEFL Score"], color = 'pink')
plt.xlabel("TOEFL Score")
plt.ylabel("GRE Score")
plt.title("GRE Score VS TOEFL Score")
plt.show()
```

```python
#boxplot
sns.boxplot(x = "CGPA", data = dataset, showmeans="True", color = "pink")
plt.title('Box Plot of CGPA')
```

```python
#boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x=dataset['University Rating'], y=dataset['TOEFL Score'], color = 'pink')
plt.title('University Rating vs TOEFL Score')
plt.show()
```

```python
#pie chart
plt.figure(figsize=(6,6))
research = dataset['Research'].value_counts()
plt.pie(research, labels=['Research (1)', 'No Research (0)'], autopct='%1.1f%%', colors=['#FF91A4', '#FFBAD2',])
plt.title('Proportions of Research')
plt.show()
```

```python
#line chart
plt.plot(dataset["Serial No."], dataset["CGPA"], color = 'pink')
plt.xlabel('Serial No.')
plt.ylabel('CGPA')
plt.title('')
plt.show()
```
